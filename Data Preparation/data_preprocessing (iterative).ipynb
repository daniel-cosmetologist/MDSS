```python
import pandas as pd
```

```python
df = pd.read_csv('/home/djden/Development/Science/FateOfFat/MDSS/Initial Data/db_nl_preprocessed.csv')
# df.drop(df.columns[0], axis=1, inplace=True)
df
```

```python
X = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer(max_iter=10, random_state=42)

X_imputed = imputer.fit_transform(X)
X_imputed
```

```python
X_imputed_df = pd.DataFrame(X_imputed, columns=X.columns)
X_imputed_df['target'] = y.reset_index(drop=True)
X_imputed_df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_imputed_scaled = scaler.fit_transform(X_imputed_df.drop('target', axis=1))
X_imputed_scaled_df = pd.DataFrame(X_imputed_scaled, columns=X.columns)
X_imputed_scaled_df['target'] = X_imputed_df['target']
```
